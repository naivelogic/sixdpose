```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from PIL import ImageMath
import sys
sys.path.append('..')

img = Image.open('../data/demo/0000.png')
print(img)

# convert to RGB
if img.mode != 'RGB':
    if img.mode == 'I':
        from PIL import ImageMath
        img = ImageMath.eval('im/256', {'im': img})
    img = img.convert('L')
    
    #img = img.convert('RGB')

    np_frame = np.array(img)
    
np_frame = np.array(img)
print(np_frame.shape)
imgplot = plt.imshow(np_frame)
imgplot.axes.get_xaxis().set_visible(False)
imgplot.axes.get_yaxis().set_visible(False)
```

```python
def pil_loader(path):
    with open(path, 'rb') as f:
        img = Image.open(f)
        
        if img.mode == 'I':
            img = ImageMath.eval('im/256', {'im': img})
            
        return img.convert('RGB')
```

```python
img = pil_loader('../data/demo/0000.png')

np_frame = np.array(img)
print(np_frame.shape)
imgplot = plt.imshow(np_frame)
```
