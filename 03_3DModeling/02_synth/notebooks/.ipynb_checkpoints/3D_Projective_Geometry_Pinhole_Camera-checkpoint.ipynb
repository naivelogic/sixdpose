### IPinhole Camera Model


__Pinhole cama parameters__ are represented in a 4-by-3 matrix called the __camera matrix__. This matrix maps the 3D world scene into the image plane. The __calibration algorithm__ calculates the camear matrix using the `extrinsic` and `intrinsic` parameters. 
- __extrinsic parameters__ - represent the locaiton of the camera in the 3D scene in a __ridgid transformation__ from the 3D world cooridnate syste to the 3D cameras cooridinate system.
- __intrinsic parameters__ - represent the __optical center__ and __focal length__ of the camera that then is a __projective transformation__ from the 3D camera cooridnates into the 2D image cooridinates


![](https://www.mathworks.com/help/vision/ug/calibration_camera_matrix.png) ![](https://www.mathworks.com/help/vision/ug/calibration_coordinate_blocks.png)

_https://www.mathworks.com/help/vision/ug/camera-calibration.html_




IN which we define the __intrinsic matrix as__ for the camera model as the followign which provides a geometric property of the camera. 

$K = \left ( 
                \begin{array}{ c c c}
                f_x & s   & x_0 \\
                 0  & f_y & y_0 \\
                 0  & 0   & 1 \\
                \end{array}
            \right )$
            
            
Essentially all we are is calcualting the perspective projection mmodel by the ideal pinhole camera.

-> more details: http://ksimek.github.io/2013/08/13/intrinsic/

```python
K = array([[5,0,0],[0,5,0],[0,0,1]])
```

```python
def plotpoints(k):
    
    th = k*pi/16
    XYZ=array([(5*i,5*j*cos(th),50+5*j*sin(th)) \
               for i in range(-2,3) for j in range(-2,3)]).T
    
    xys = dot(K,XYZ)
    xy = 1.0*xys[:2,:]/xys[-1,:]
    plot(xy[0], xy[1], 'o')
    axis('equal')
```

```python
plotpoints(6)
show()
```
