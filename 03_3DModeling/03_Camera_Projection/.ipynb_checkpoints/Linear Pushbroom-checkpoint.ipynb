## Sweep Generator

Next we geneerate the surfaces by wraping the images based on the referece camera based on each pixel and depth. We will need to ocompute the sweep volume which requires a transofmraiton fo the camera points.

```python
sweep_iter = 256 # number of sweeps of planes
depth_inverse = np.linspace(1/0.8, 1/8.0, sweep_iter)
```

```python
def wvar(a, w):
    av = np.average(a, weights=w, axis=0)
    av=np.expand_dims(av,axis=0)
    var = np.average((a - av) ** 2, weights=w,axis=0)
    return var

def compute_relative_pose(cam_pose, ref_pose):
    relative_pose = np.zeros((3, 4), dtype=np.float64)

    Ri,Rr=cam_pose[:,:-1],ref_pose[:,:-1]
    ti,tr=cam_pose[:,-1:],ref_pose[:,-1:]
    relative_pose[:,:-1]=Ri@Rr.T
    relative_pose[:, -1:]=ti-Ri@Rr.T@tr
    return relative_pose

def get_plane_sweep_homographies(K, relative_pose, inv_depths):
    r=relative_pose[:,:-1]
    t=relative_pose[:,-1:]
    n=np.array([[0,0,1]])
    invk=np.linalg.inv(K)
    le=np.expand_dims(K@r@invk,axis=0)
    inv_depths=np.expand_dims(inv_depths,axis=(1,2))
    homographies=K@((t@n)*inv_depths)@invk + le
    
    return homographies

#https://github.com/tallan007/unccv_image_stitching
```

```python

#https://github.com/tallan007/unccv_image_stitching
def Genearte_sweeps(images, ref_pose, K, inv_depths, img_hw):
    D = len(inv_depths)
    H, W = img_hw
    N=len(images)
    ps_volume = np.zeros((D, H, W), dtype=np.float32)
    ps_all = np.zeros([N,H,W,3], dtype=np.float32)
    accum_count = np.zeros((D, H, W), dtype=np.uint8)
    mask=np.zeros([N,H,W,3],dtype=np.uint8)
    extras = []
    
    H = np.zeros([N,D,3,3])
    
    for i in range(N):
        #relative_pose = compute_relative_pose(images[i].qvec.reshape(3,3), ref_pose)
        #relative_pose = compute_relative_pose(images[i].vec.reshape(3,3), ref_pose)
        relative_pose = compute_relative_pose(images[i].extrinsics, ref_pose)
        H[i] = get_plane_sweep_homographies(K, relative_pose, np.array(inv_depths))
    extras.append(H)
    boraderValue=-1
    for j in range(0,D):
        for i in range(N):
            ps_all[i]=cv2.warpPerspective(images[i].image,np.linalg.inv(H[i][j]),img_hw[::-1],
                                          borderValue=(boraderValue,boraderValue,boraderValue))
        mask[:, :, :, :] = (ps_all != boraderValue)
        accum_count[j]=np.sum(mask[:,:,:,0],axis=0)
        ps_volume[j]=np.mean(wvar(ps_all,w=mask),axis=2)
    return ps_volume, accum_count, extras
```

```python
ps_volume, accum_count, extras = Genearte_sweeps(img, test_pose, K, depth_inverse,(im_height, im_width))
```

## Visualize Point Could of 3D Model

```python
def unproject_depth_map(image, inv_depth_image, K, mask=None):
    H, W = image.shape[:-1]
    xx, yy = np.meshgrid(np.arange(W), np.arange(H))
    homocoor = np.concatenate([xx.reshape(-1, 1), yy.reshape(-1, 1), np.ones([H * W, 1])], axis=1)

    xyz = np.zeros([H * W, 3], dtype=np.float64)
    rgb = np.zeros([H * W, 3], dtype=np.float64)

    depth = inv_depth_image[yy.reshape(-1), xx.reshape(-1)]
    xyz[:, -1] = -depth / (homocoor @ np.linalg.inv(K).T)[:, -1]
    xyz[:, :-1] = (homocoor @ np.linalg.inv(K).T)[:, :-1]

    rgb[:, :] = image[yy.reshape(-1), xx.reshape(-1)]

    if (mask is not None):
        coormask = mask[yy.reshape(-1), xx.reshape(-1)]
        coormask = coormask.astype(np.bool)
        xyz = xyz[coormask]
        rgb = rgb[coormask]
    return xyz, rgb
```

```python
def rgb2hex(rgb):
    """Converts color representation into hexadecimal representation for K3D

    Args:
        rgb (np.ndarray): (N, 3) array holding colors

    Returns:
        hex (np.ndarray): array (N, ) of size N, each element indicates the
          color, e.g. 0x0000FF = blue
    """
    rgb_uint = (rgb * 255).astype(np.uint8)
    hex = np.sum(rgb_uint * np.array([[256 ** 2, 256, 1]]),
                 axis=1).astype(np.uint32)
    return hex
```
